```python
my_db = mysql.connector.connect(host="localhost",user="root",passwd="mysql",database="w3schools")
```

```python
my_cursor=my_db.cursor()
```

```python
my_cursor.execute("show tables")
```

```python
for i in my_cursor:
    print(i)
```

```python
my_cursor.execute("SELECT DISTINCT orders.orderid AS OrderID,orders.orderdate AS OrderDate,customers.customerid AS CustomerID,customername,contactname,city,country FROM customers INNER JOIN orders ON customers.customerid = orders.customerid;")
data=my_cursor.fetchall()
print(data)
```

```python
df=pd.DataFrame(data,columns =["OrderID","OrderDate","CustomerID","CustomerName","ContactName","City","Country"])
```

```python
df
```

```python
from pandas.plotting import table
```

```python
plt.figure()

# table
plt.subplot(121)

cell_text = []
for row in range(len(df)):
    cell_text.append(df.iloc[row])

plt.table(cellText=cell_text, colLabels=df.columns, loc='center')
plt.axis('off')
figure(figsize=(4000, 5000), dpi=50000000)

plt.show()
```
